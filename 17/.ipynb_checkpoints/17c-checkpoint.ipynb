```python
# importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor

# Create the dataset
rng = np.random.RandomState(1)
X = np.linspace(0, 6, 100)[:, np.newaxis]
y = np.sin(X).ravel() + np.sin(6 * X).ravel() + rng.normal(0, 0.1, X.shape[0])

# Fit regression model
regr_1 = DecisionTreeRegressor(max_depth=1)

regr_2 = DecisionTreeRegressor(max_depth=2)

regr_3 = DecisionTreeRegressor(max_depth=20)

regr_1.fit(X, y)
regr_2.fit(X, y)
regr_3.fit(X, y)

# Predict
y_1 = regr_1.predict(X)
y_2 = regr_2.predict(X)
y_3 = regr_3.predict(X)

# Plot the results
plt.figure()
plt.scatter(X, y, c="k", label="training samples")
plt.plot(X, y_1, c="g", label="max_depth=1", linewidth=2)
plt.plot(X, y_2, c="r", label="max_depth=2", linewidth=2)
plt.plot(X, y_3, c="b", label="max_depth=20", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_score, recall_score, f1_score
X, y = load_iris(return_X_y=True)
model = DecisionTreeClassifier()
# result = cross_val_score(DecisionTreeClassifier(), X, y)
# result
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=None)
y_pred = DecisionTreeClassifier().fit(X_train, y_train).predict(X_test)

# weighted average precision
print(precision_score(y_test, y_pred, average='weighted'))

# weighted average recall
print(recall_score(y_test, y_pred, average='weighted'))

# precision 和 recall 的调和平均数
print(f1_score(y_test, y_pred, average='weighted'))
```

```python
import numpy as np
clf = DecisionTreeClassifier().fit(X_train, y_train)
y_prob = clf.predict_proba(X_test)
np.isclose(np.argmax(y_prob, axis=1), clf.predict(X_test)).all()
```

## 决定系数 (coefficient of determination): $r^2$ 


在统计学中用于度量因变量的变异中可由自变量解释部分所占的比例，以此来判断统计模型的解释力。 <br>

对于简单线性回归而言，决定系数为样本相关系数的平方。<br>
