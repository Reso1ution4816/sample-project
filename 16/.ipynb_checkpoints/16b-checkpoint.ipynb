为什么使用 one-hot encoding? <br>

如果使用 `sklearn.preprocessing.LabelEncoder`, `Czech Republic` 会变成 `10`, `India` 会变成 `20`. <br>

`10` 和 `20` 的平均数是 `15`, 在encoding里对应的是 `Germany`. <br>

但显然由于类型(categorical)变量没有顺序, `India` 和 `Czech Republic` 并不能求平均数, 即便有, 其"平均数" `Germany` 也无法理解

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt

def test_and_plot(X, y, regr, random_state=None):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.33, random_state=42)
    y_pred = regr.fit(X_train, y_train).predict(X_test)
    result = pd.DataFrame(data={'true':y_test, 'predicted':y_pred, 'residual':y_test-y_pred})
    result['residual'].hist()
    print('r_squared=%s'% r2_score(y_test, y_pred), 
          '\nmean_squared_error=%s'% mean_squared_error(y_test, y_pred),
          '\nmean_absolute_error=%s'% mean_absolute_error(y_test, y_pred))
    plt.show()

regr = LinearRegression()
test_and_plot(ndX, ndy, regr, random_state=42)
```

```python
regr = DecisionTreeRegressor(max_depth=5)
test_and_plot(ndX, ndy, regr, random_state=42)
```

## Log Loss

This is the multi-class version of the [LogarithmicLoss](https://www.kaggle.com/wiki/LogarithmicLoss) metric. Each observation is in one class and for each observation, you submit a predicted probability for each class. The metric is negative the log likelihood of the model that says each test observation is chosen independently from a distribution that places the submitted probability mass on the corresponding class, for each observation.

$$log loss = -\frac{1}{N}\sum_{i=1}^N\sum_{j=1}^My_{i,j}\log(p_{i,j})$$

where N is the number of observations, M is the number of class labels, \\(log\\) is the natural logarithm, \\(y_{i,j}\\) is 1 if observation \\(i\\) is in class \\(j\\) and 0 otherwise, and \\(p_{i,j}\\) is the predicted probability that observation \\(i\\) is in class \\(j\\).

Both the solution file and the submission file are CSV&#39;s where each row corresponds to one observation, and each column corresponds to a class. The solution has 1&#39;s and 0&#39;s (exactly one &quot;1&quot; in each row), while the submission consists of predicted probabilities.

The submitted probabilities need not sum to 1, because they will be rescaled (each is divided by the sum) so that they do before evaluation.

(Note: the actual submitted predicted probabilities are replaced with \\(max(min(p,1-10^{-15}),10^{-15})\\).)

Note: In this example &#39;publicScore&#39; is the score for only the rows marked as &#39;Public&#39;. The observation with ID of 7 is ignored in the calculation of 1.881797068998267 (and \\(N=6\\)).


```python
import numpy as np

def multiclass_log_loss(y_true, y_pred, eps=1e-15):
    """Multi class version of Logarithmic Loss metric.
    https://www.kaggle.com/wiki/MultiClassLogLoss

    Parameters
    ----------
    y_true : array, shape = [n_samples]
            true class, intergers in [0, n_classes - 1)
    y_pred : array, shape = [n_samples, n_classes]

    Returns
    -------
    loss : float
    """
    predictions = np.clip(y_pred, eps, 1 - eps)

    # normalize row sums to 1
    predictions /= predictions.sum(axis=1)[:, np.newaxis]

    actual = np.zeros(y_pred.shape)
    n_samples = actual.shape[0]
    actual[np.arange(n_samples), y_true.astype(int)] = 1
    vectsum = np.sum(actual * np.log(predictions))
    loss = -1.0 / n_samples * vectsum
    return loss

a = np.asarray([0,0,0,1,1,1])
b = np.asarray([[0.5,0.5],[0.1,0.9],[0.01,0.99],[0.9,0.1],[0.75,0.25],[0.001,0.999]])
print('expected=1.881797068998267\ncalculated=', multiclass_log_loss(a, b))

from sklearn.metrics import log_loss
print('can we replace it with sklearn\'s log loss function?\nsklearn.metrics.log_loss=', log_loss(a,b))
```

## 决策树可以输出概率预测

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss, accuracy_score

X, y = load_iris(return_X_y=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)

y_pred_proba = DecisionTreeClassifier(max_depth=1).fit(X_train, y_train).predict_proba(X_test)
y_pred_proba.shape
```

```python
y_pred_proba
```

### log loss

```python
log_loss(y_test, y_pred_proba)
```

### accuracy

```python
accuracy_score(y_test, np.argmax(y_pred_proba, axis=1))
```
