# 回归问题

给定不同地区的收入, 学校师生比等等共13列数据, 和该地区对应的房价中位数, 让算法概括如何根据这些数据来估算一个地区的房价
* 模型定义为一个函数 <br><br>
  输入是$m$个地区的信息$\textbf{X}$, 表示为一个$m\times 13$矩阵 <br><br>
  输出是$m$个地区的房价中位数$\textbf{y}$, 表示为一个长度为$m$的向量 : <br><br>
<br><br>
   \begin{equation}
   f(\textbf{X}):=\{\textbf{X} \mapsto \textbf{y} \}\\
   \textbf{X} \in \mathbb{R}^{m\times 13},
   \textbf{y} \in \mathbb{R}^m
   \end{equation}
<br><br>
<br><br>

* 假设 $\textbf{y}$是真实的$m$个房价中位数, 模型预测的房价中位数为 $f(\textbf{X}_i), i = 1,2,...,m$ 我们想要最小化两者的差距. 
<br><br>
     \begin{equation}
      \epsilon = |\textbf{y} - f(\textbf{X})| = \sum_{i=1}^{m} |\textbf{y} - f(\textbf{X})|
     \end{equation}
<br><br>
* 优化: 我们希望找到误差最低的模型 $f$
<br><br>
 \begin{equation}
     f = argmin(\epsilon) = argmin \sum_{i=1}^{m} |\textbf{y} - f(\textbf{X})|
 \end{equation}
 <br><br>
<br>
* 但由于优化(最大化/最小化)$|\textbf{u}|$ 没有显式解 (closed form solution), 通常大家会优化 $||\textbf{u}||_2$. 也就是
<br><br>
 \begin{equation}
     f = argmin \sum_{i=1}^{m} ||\textbf{y} - f(\textbf{X})||_2
 \end{equation}
 <br><br>
<br>

# 线性回归(Linear Regression)模型, 也叫 Ordinary Least Square (OLS) 模型

<br><br>
 \begin{equation}
     \textbf{y} = \textbf{w}_{0} * \underbrace{[1,...,1]}_{m} + \textbf{w}_1*\textbf{X}_{:,1} + \textbf{w}_2*\textbf{X}_{:,2} + ... + \textbf{w}_n*\textbf{X}_{:,n}\\
     \textbf{y} = (\underbrace{[1,...,1]^T}_{m} | \textbf{X} ) \textbf{w}\\
     \text{where }\textbf{X} \in \mathbb{R}^{m\times n}, \textbf{y} \in \mathbb{R}^m, \textbf{w} \in \mathbb{R}^{n+1}\\
     \quad\\
     \text{note } (\textbf{A} | \textbf{B}) \text{is known as augmentation, you may see it as column stack}\\
     \text{hence } (\underbrace{[1,...,1]^T}_{m} | \textbf{X} ) \in \mathbb{R}^{m\times (n+1)}
 \end{equation}
 <br><br>
<br>
more on matrix augmentation<br>
https://en.wikipedia.org/wiki/Augmented_matrix

```python
import numpy as np
X = np.array([[1, 5], [3, 2], [6, 1]])
y = np.array([2, 3, 4])
w = np.array([2.4285714285714288, 0.28571429, -0.14285714])

ones = np.ones(X.shape[0])
augX = np.column_stack([ones, X])
y_pred = augX.dot(w)
np.isclose(y, y_pred)
```

## $argmin_{\hat{\textbf{w}}} ||\textbf{X}\hat{\textbf{w}} - \textbf{y}||_2^2$ 的显式解

注意, 为方便起见, 下面的 $\textbf{X}$ 指代上方的 $(\underbrace{[1,...,1]^T}_{m} | \textbf{X} )$ <br>

<br>我们想要求解一个能使模型拟合程度最高的 $\textbf{w}$<br>

假设存在一个 $\hat{\textbf{w}}$ 满足 $\textbf{X}\hat{\textbf{w}}\approx \textbf{y}$, 模型拟合程度可以表示为 $||\textbf{X}\hat{\textbf{w}} - \textbf{y}||_2^2$, 那么$\hat{\textbf{w}}$可以表示为:<br>

<br><br>
 \begin{equation}
     \hat{\textbf{w}} = argmin(J(\hat{\textbf{w}}))     = argmin ||\textbf{X}\hat{\textbf{w}} - \textbf{y}||_2^2
 \end{equation}
 <br><br>
<br>
$argmin_{\hat{\textbf{w}}} ||\textbf{X}\hat{\textbf{w}} - \textbf{y}||_2^2$ 具有显式解:
<br><br>
 \begin{equation}
     \hat{\textbf{w}} = (\textbf{X}^T\textbf{X})^{-1}\textbf{X}^T\textbf{y}
 \end{equation}
 <br><br>
<br>

```python
import numpy as np


class MyLinearRegression:
    def __init__(self):
        self.w = None

    @staticmethod
    def ones_augment_to_left(X):
        X = np.array(X)
        ones = np.ones(X.shape[0])
        return np.column_stack([ones, X])

    def fit(self, X_train, y_train):
        X = self.ones_augment_to_left(X_train)
        y = np.array(y_train)

        product = np.dot(X.T, X)
        theInverse = np.linalg.inv(product)
        self.w = np.dot(np.dot(theInverse, X.T), y)
        return self

    def predict(self, X_test):
        X_test = np.array(X_test)
        return self.ones_augment_to_left(X_test).dot(self.w)

    def predict_alternative(self, X_test):
        predictions = []
        for i in X_test:
            components = self.w[1:] * i
            predictions.append(sum(components) + self.w[0])
        return predictions
```

```python
# 测试
import numpy as np

mlr = MyLinearRegression()

X = [[1, 5], [3, 2], [6, 1]]
y = [2, 3, 4]
y_pred = mlr.fit(X, y).predict(X)
print('Am I correct? \n', np.isclose(y, y_pred).all())

'''
如果实现正确, 会输出
Am I correct? 
 True
'''
```

## Linear Regression In Action

有效的线性回归模型须满足以下四个条件:

1. depdent variable $\textbf{y}$ 和 independent variables $\textbf{X}_{:,1}, ..., \textbf{X}_{:,n}$ 满足线性关系, independent variables 之间彼此独立

1. 残差之间没有相关性

1. 回归线两侧残差的variance相等或很相近 (homoscedasticity的非正式解读)

1. 残差服从正态分布, 如果放松限制, 非正式地讲可以认为残差的直方图应是钟形曲线

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# Load the diabetes dataset

diabetes = datasets.load_diabetes()
X, y = diabetes.data, diabetes.target

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.33, 
                                                    random_state=42)

# Create linear regression object
mlr2 = MyLinearRegression()

# Train the model using the training sets
mlr2.fit(X_train, y_train)

# Make predictions using the testing set
y_pred = mlr2.predict(X_test)

# The coefficients
w = mlr2.w
print('w: \n', w)

# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, y_pred))

# Explained variance score: 1 is perfect prediction
print('Variance score (r squared, r^2): %.2f' % r2_score(y_test, y_pred))
```

## 决定系数 (coefficient of determination): $r^2$ 

`sklearn.metrics.r2_score(y_true, y_pred, sample_weight=None, multioutput='uniform_average')`

R^2 (coefficient of determination) regression score function.

Best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse). A constant model that always predicts the expected value of y, disregarding the input features, would get a R^2 score of 0.0.

http://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html

在统计学中用于度量因变量的变异中可由自变量解释部分所占的比例，以此来判断统计模型的解释力。 <br>

对于简单线性回归而言，决定系数为样本相关系数的平方。<br>

来源:<br> 
https://zh.wikipedia.org/wiki/%E5%86%B3%E5%AE%9A%E7%B3%BB%E6%95%B0 <br> 

若中文维基无法访问: <br> 
https://en.wikipedia.org/wiki/Coefficient_of_determination <br> 

额外阅读:<br>
http://blog.minitab.com/blog/adventures-in-statistics-2/regression-analysis-how-do-i-interpret-r-squared-and-assess-the-goodness-of-fit

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(7,14), dpi=120)

for i in range(X_train.shape[1]):
    y_pred3 = MyLinearRegression().fit(
        X_train[:, i], y_train).predict(
        X_test[:, i])
    
    ax = fig.add_subplot(5, 2, i + 1)
    
    ax.set_xlabel('X[:, %d]' % i)
    
    ax.set_ylabel('y')
    
    ax.scatter(X_train[:, i], y_train, 
               color='black', marker='^', 
               alpha=0.5, label='Train Data')
    
    ax.scatter(X_test[:, i], y_test, 
               color='black', 
               alpha=0.5, label='Test Data')
    
    ax.plot(X_test[:, i], y_pred3, 
            color='blue', 
            linewidth=3, label='Fitted Line')
    
    ax.set_title('MSE=%.2f, r^2=%.2f'% (
                mean_squared_error(y_test, y_pred3), 
                r2_score(y_test, y_pred3)
                )
            )

fig.tight_layout()
plt.legend()
plt.show()
```

## 残差 Residuals

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(5, 2, figsize=(7,14), dpi=120)
for i, ax in enumerate(np.ravel(axes)):
    y_pred3 = MyLinearRegression().fit(
        X_train[:, i], y_train).predict(
        X_test[:, i])
    sns.kdeplot(y_test - y_pred3, ax=ax)
    ax.set_xlabel('Residuals')
    ax.set_ylabel('Density Estimation')
    ax.set_title('X[:, %d] MSE=%.2f, r^2=%.2f'% (
                i,
                mean_squared_error(y_test, y_pred3), 
                r2_score(y_test, y_pred3)
                )
            )
fig.tight_layout()
plt.show()
```

# 使用 `sklearn` 做线性回归

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
# Load the diabetes dataset

diabetes = datasets.load_diabetes()
X, y = diabetes.data, diabetes.target

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.33, 
                                                    random_state=42)

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X_train, y_train)

# Make predictions using the testing set
y_pred = regr.predict(X_test)


# The coefficients
w_1_to_n = regr.coef_
w0 = regr.intercept_
print('w_0: \n', w0)
print('w_1 to w_n: \n', w_1_to_n)

# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, y_pred))

# Explained variance score: 1 is perfect prediction
print('Variance score (r squared, r^2): %.2f' % r2_score(y_test, y_pred))
```
